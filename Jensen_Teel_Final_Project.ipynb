```r
library(tidyverse)
library(glue)
library(ggplot2)
library(readxl)
```

```r
videogames <- read_csv("https://raw.githubusercontent.com/Jteel2020/datasets/main/Video_Game_Sales_as_of_Jan_2017.csv")
```

```r
glimpse(videogames)
```

```r
print(videogames)
```

```r
data = videogames
video_games <- data %>% select(Year_of_Release, Global_Sales) %>% 
arrange(desc(Year_of_Release))
```

```r
ggplot(video_games, aes(x = Year_of_Release)) + geom_bar(fill = "red", color = "black") +
  labs(title = "Global Video Game Sales per Year", x = "Year", y = "Global Sales") +
  theme(plot.title = element_text(hjust = .5, face = "bold"), axis.title = element_text(face = "bold"))
```

```r
datas = videogames
video_games_2 <- datas %>% select(Publisher, Critic_Score, Rating) %>%
filter(Publisher %in% c("Ubisoft", "Nintendo", "Electronic Arts", "Activision"))
```

```r
ggplot(video_games_2, aes(x = Publisher, y = Critic_Score, fill = Publisher)) + geom_boxplot(outlier.shape = NA) +
 labs(title = "Critic Scores based on Publisher", x = "Publisher", y = "Critic Score") +
  theme(plot.title = element_text(hjust = .5, face = "bold"), axis.title = element_text(face = "bold")) 
```
